#Introduction
Tokenization is one of the first step in any NLP pipeline. Tokenization is nothing but splitting the raw text into small chunks of words or sentences, called **tokens**. **If the text is split into words, then its called as 'Word Tokenization'** and if **it's split into sentences then its called as 'Sentence Tokenization'**.

Generally 'space' is used to perform the word tokenization and characters like 'periods, exclamation point and newline char are used for Sentence Tokenization. We have to choose the appropriate method as per the task in hand. While performing the tokenization few characters like spaces, punctuations are ignored and will not be the part of final list of tokens.

###1.Tokenization Using Python's Inbuilt Method - Split()

```python
#Spliting the paragraph into sentences.
paragraph="""My dear young friends, dream, dream, dream. Dreams transform into thoughts and thoughts result in action. You have to dream before your dreams can come true. You should have a goal and a constant quest to acquire knowledge. Hard work and perseverance are essential. Use technology for the benefit of humankind and not for its destruction. The ignited mind of the youth is the most powerful resource on the earth, above the earth, and under the earth. When the student is ready, the teacher will appear. Aim high, dream big, and work hard to achieve those dreams. The future belongs to the young who have the courage to dream and the determination to realize those dreams. Remember, small aim is a crime; have great aim and pursue it with all your heart."""
tokens= paragraph.split('.')
tokens
```

###2.Tokenization Using Regular Expressions(RegEx)
**RegEx Function:** The '*re* ' module offers a set of function that allows us to search a string for a match.

```python
#Sentence tokenization
import re
paragraph="""My dear young friends, dream, dream, dream. Dreams transform into thoughts and thoughts result in action. You have to dream before your dreams can come true. You should have a goal and a constant quest to acquire knowledge. Hard work and perseverance are essential. Use technology for the benefit of humankind and not for its destruction. The ignited mind of the youth is the most powerful resource on the earth, above the earth, and under the earth. When the student is ready, the teacher will appear. Aim high, dream big, and work hard to achieve those dreams. The future belongs to the young who have the courage to dream and the determination to realize those dreams. Remember, small aim is a crime; have great aim and pursue it with all your heart."""
sentences= re.compile('[.!?]').split(paragraph)
sentences
```

```python
#Word tokenization
import re
paragraph="""My dear young friends, dream, dream, dream. Dreams transform into thoughts and thoughts result in action. You have to dream before your dreams can come true. You should have a goal and a constant quest to acquire knowledge. Hard work and perseverance are essential. Use technology for the benefit of humankind and not for its destruction. The ignited mind of the youth is the most powerful resource on the earth, above the earth, and under the earth. When the student is ready, the teacher will appear. Aim high, dream big, and work hard to achieve those dreams. The future belongs to the young who have the courage to dream and the determination to realize those dreams. Remember, small aim is a crime; have great aim and pursue it with all your heart."""
words= re.compile('\W+').split(paragraph)
print(words)
```

###3.Tokenization Using NLTK

```python
# Library import
import nltk
nltk.download()
```

```python
#Sentence tokenization
paragraph= """ My dear young friends, dream, dream, dream.
Dreams transform into thoughts and thoughts result in action.
You have to dream before your dreams can come true.
 You should have a goal and a constant quest to acquire knowledge.
  Hard work and perseverance are essential.
   Use technology for the benefit of humankind and not for its destruction.
   The ignited mind of the youth is the most powerful resource on the earth, above the earth, and under the earth.
   When the student is ready, the teacher will appear.
    Aim high, dream big, and work hard to achieve those dreams.
     The future belongs to the young who have the courage to dream and the determination to realize those dreams.
      Remember, small aim is a crime; have great aim and pursue it with all your heart."""
sentences= nltk.sent_tokenize(paragraph)
sentences
```

```python
#Word tokenization
words=nltk.word_tokenize(paragraph)
print(words)
```

###4.Tokenization using Spacy


1.   Spacy:spaCy is an open-source software library for advanced natural language processing, written in the programming languages Python and Cython
2.   In spaCy we create language model object, which then used for word and sentence tokenization.

3.   Syntax to install spaCy library and English model is as below

```python
#Setence Tokenizer
import spacy
# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")
# Process the text with SpaCy
doc = nlp(paragraph)
# Extract sentences
sentences = list(doc.sents)
# Print each sentence
for sentence in sentences:
    print(sentence.text)
```

```python
#Word Tokenization
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(paragraph)

word_tokens = [token.text for token in doc]

print("Word Tokens:", word_tokens)
```

###5.Tokenization using Keras

1.   Keras is opensource neural network library written in python.

2.   It is easy to use and it is capable of running on top of TensorFlow, Microsoft Cognitive Toolkit, R, Theano, or PlaidML

3.   To perform word tokenization we use the text_to_word_sequence() method from the keras.preprocessing.text class
By default, this function automatically does 3 things:



*   Splits words by space (split=” “).
*   Filters out punctuation (filters=’!”#$%&()*+,-./:;<=>? [\]^_`{|}~\t\n’).
*   Converts text to lowercase (lower=True).







```python
!pip install keras
```

```python
#sentence tokenization
from keras.preprocessing.text import text_to_word_sequence

text_to_word_sequence(paragraph, split= ".", filters="!.\n")
```

```python
#Word Tokenization
from keras.preprocessing.text import text_to_word_sequence
tokens = text_to_word_sequence(paragraph)
print(tokens)
```

# Introduction:

In Natural Language Processing (NLP), POS stands for Part of Speech. It refers to the grammatical categories into which words are classified based on their syntactic roles within sentences. Understanding the POS of words is essential for various NLP tasks, as it helps in parsing sentences, understanding context, and improving the accuracy of language models.

Common Parts of Speech:

**Noun (NN):** Represents a person, place, thing, or idea (e.g., dog, London, love).

**Verb (VB): **Indicates an action, event, or state (e.g., run, is, think).

**Adjective (JJ):** Describes or modifies a noun (e.g., blue, quick, tall).

**Adverb (RB): **Modifies a verb, an adjective, or another adverb (e.g., quickly, very, well).

**Pronoun (PRP):** Replaces a noun (e.g., he, she, it).

**Preposition (IN):** Shows the relationship between a noun (or pronoun) and other words in a sentence (e.g., in, on, at).

**Conjunction (CC):** Connects words, phrases, or clauses (e.g., and, but, or).

**Determiner (DT):** Introduces a noun and provides context (e.g., the, a, some).

**Interjection (UH):** Expresses strong emotion or reaction (e.g., wow, oh, ouch).

**Importance of POS Tagging in NLP:**

1. Text Parsing: Helps in breaking down a sentence into its grammatical components, making it easier to analyze the structure.
2. Named Entity Recognition (NER): Identifying names of people, places, etc., often relies on accurate POS tagging.
3. Sentiment Analysis: Understanding whether a word is an adjective, verb, or noun can help determine the sentiment expressed.
4. Machine Translation: Correctly translating a sentence requires an understanding of the roles each word plays.
5. Speech Recognition: Identifying and correctly interpreting words based on their function.

**POS Tagging:**
POS tagging is the process of labeling each word in a sentence with its corresponding part of speech. This can be done using rule-based methods, statistical models, or machine learning techniques, such as Hidden Markov Models (HMMs), Conditional Random Fields (CRFs), or deep learning models.


# POS Techniques:
Several techniques have been developed for POS tagging, ranging from simple rule-based approaches to more sophisticated machine learning models. Here are some of the common techniques:

1. Rule-Based POS Tagging: Rule-based taggers use a set of hand-crafted linguistic rules to assign POS tags. These rules consider the word itself, its suffixes or prefixes, and its context within the sentence.

2. Statistical POS Tagging: Statistical taggers use probabilities derived from large annotated corpora to determine the most likely POS tag for a word in a given context. Common statistical models include Hidden Markov Models (HMMs) and Maximum Entropy models.




*   HMM: A Hidden Markov Model (HMM) is a statistical model used to represent systems or processes that are assumed to follow the Markov property, but with the added complexity that the states of the system are hidden or unobservable. HMMs are widely used in various fields, including natural language processing, speech recognition, bioinformatics, and more.

Key Concepts of HMM:

1. **Markov Property:**  The Markov property states that the probability of transitioning to any particular state depends only on the current state and not on the sequence of states that preceded it. In other words, the future state depends only on the present state and not on the past states.

2. **Hidden States:** In an HMM, the system is assumed to be in one of a finite number of states at any given time, but these states are not directly observable. Instead, they are hidden. What is observable is some output or observation that is probabilistically related to these hidden states.

3. **Observations:** Each hidden state is associated with a probability distribution over possible observations. When the system is in a particular state, it generates an observation according to this distribution. Thus, the sequence of observed events is generated by an underlying sequence of hidden states.

## Rule based POC tagging:


```python
import re

# Define a simple set of rules for POS tagging
def pos_tag(word):
    # Rule 1: If the word ends in 'ing', it's a verb (gerund form)
    if re.search(r'ing$', word):
        return 'VERB'
    # Rule 2: If the word ends in 'ed', it's a verb (past tense)
    elif re.search(r'ed$', word):
        return 'VERB'
    # Rule 3: If the word ends in 'ly', it's an adverb
    elif re.search(r'ly$', word):
        return 'ADVERB'
    # Rule 4: If the word ends in 's', it's a noun (plural form)
    elif re.search(r's$', word):
        return 'NOUN(Plural)'
    # Rule 5: If the word is capitalized, assume it's a proper noun
    elif word[0].isupper():
        return 'PROPERNOUN'
    # Rule 6: Default rule, assume the word is a noun
    else:
        return 'NOUN'

# Tokenize the sentence (simple split on spaces)
tokens = paragraph.split()

# Apply the POS tagger
tagged_sentence = [(word, pos_tag(word)) for word in tokens]

list(tagged_sentence)
```

## Statistical POS Tagging:

```python
#Hidden Markov Model(HMM)
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import hmm
from nltk.corpus import treebank

# Ensure necessary NLTK data is downloaded
nltk.download('treebank')
nltk.download('universal_tagset')
nltk.download('punkt')
```

```python
# Tokenize sentence properly
sentence = "The quick brown fox jumps over the lazy dog."

tokens = word_tokenize(sentence)

# Load more data for training
train_data = treebank.tagged_sents(tagset='universal')[:5000]

# Train the HMM POS tagger
trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(train_data)

# Test the HMM POS tagger on the tokenized sentence
tagged_sentence = hmm_tagger.tag(tokens)

print("HMM POS Tagging:")
print(tagged_sentence)
```

```python
#Maximum Entropy Model POS Tagging with NLTK

import nltk
from nltk.corpus import treebank
from nltk.tag import untag
from nltk.classify import MaxentClassifier
from nltk.tag.sequential import ClassifierBasedPOSTagger

# Ensure necessary NLTK data is downloaded
nltk.download('treebank')
nltk.download('universal_tagset')

# Load the Treebank corpus
train_data = treebank.tagged_sents(tagset='universal')[:3000]
test_data = treebank.tagged_sents(tagset='universal')[3000:]

# Train a Maximum Entropy POS tagger
maxent_tagger = ClassifierBasedPOSTagger(train=train_data, classifier_builder=MaxentClassifier.train)

# Test the Maximum Entropy POS tagger on a sentence
sentence = "They are studying machine learning and NLP".split()
tagged_sentence = maxent_tagger.tag(sentence)

print("Maximum Entropy POS Tagging:")
print(tagged_sentence)
```

## POS Tagging with Transformers (Hugging Face):

For advanced POS tagging, we can use pre-trained transformer models like BERT.

```python
!pip install transformers

```

```python
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore", message="Some weights of the model checkpoint at")

# Load a pre-trained token classification model
pos_pipeline = pipeline("token-classification", model="dbmdz/bert-large-cased-finetuned-conll03-english")

# Perform token classification on a sample text
text = "Hugging Face is creating a tool that democratizes AI."
tokens = pos_pipeline(text)

# Print the results
for token in tokens:
    print(f"Token: {token['word']}, Label: {token['entity']}, Score: {token['score']:.4f}")
```

Token: The word or part of the text that was classified.

Label: The entity type or tag assigned to the token. For instance:

I-ORG indicates that "Face" is part of an organization.

I-MISC indicates that "AI" is categorized as a miscellaneous entity.

Score: The confidence score of the prediction, ranging from 0 to 1.

```python
from transformers import BertTokenizer, BertForTokenClassification
from transformers import pipeline

# Load the pre-trained BERT model and tokenizer
model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForTokenClassification.from_pretrained(model_name)

# Load the pipeline for NER
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)

# Sample text
text = "Apple is looking at buying U.K. startup for $1 billion."

# Perform NER
entities = ner_pipeline(text)

# Print the results
for entity in entities:
    print(f"Entity: {entity['word']}, Label: {entity['entity']}, Score: {entity['score']:.4f}")
```
